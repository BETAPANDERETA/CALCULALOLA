```python
import math


#                                         ____________________________________
#                                        |                FUNCTIONS            |
#                                         ____________________________________


def showProd(matrizA,matrizB,matrizC,dec,res,i):

  for x in range(len(matrizA)):

    des = math.floor(res/10)
    res = matrizA[-x-1]*matrizB[i] + des

    if x < len(matrizC)-1:

      dDos = math.floor(res/10)*10
      u = res - dDos
      matrizC[-x-1] = u

    else:

      var = res
      dTres = math.floor(res/10)*10
      uDos = res - dTres

      matrizC[-x] = uDos

      dn = math.floor(res/10)*10
      ln = int(dn/10)
      matrizC = [ln]+matrizC

  # print(matrizC)
  return matrizC


def sumElemArr(matElem,i,y):

  extra = 0

  for x in range(len(matElem)):

    numVal = 0

    for y in range(0,y):

      numVal += matElem[x+y][i+y]
      # + matElem[x+1][i+1]+ matElem[x+2][i+2]+ matElem[x+3][i+3]

    extra = math.floor(numVal/10)

    if extra != 0:

      u = numVal - extra*10
      r = u

    else:

      r = numVal

    return [extra , r]


def eraseZero(arr):

  count = 0

  for element in arr:

    if element != 0:

      count += 1

  return count


def printFormat(array):

  charArr = []

  for i in array:

    charArr.append(str(i))

  sws = ""


  for j in charArr:

    sws +=  j + "  "

  print(sws)



#                                         ____________________________________
#                                        |                MAIN                |
#                                         ____________________________________


#  ____________________________________
# |            PRODUCTS                |
#  ____________________________________

productA = [0,0,0,0,0,0,5,5,3]
productB = [0,0,0,0,0,0,5,2,3]

#  ____________________________________
# |            VARIABLES               |
#  ____________________________________

prodCont = [0,0,0,0,0,0,0,0,0]
sums = []
arrSum = []
resF = []
d = 0
r = 0 + d


printFormat(productA)
printFormat(productB)
print("___________________________________ x \n")


for x in range(len(productA)):

  s = showProd(productA,productB,prodCont,d,r,-x-1)
  sums.append(s)


for y in range(len(sums)):

  sums[y] = sums[y]+[0]*(y)

  if y < eraseZero(productB):

    printFormat(sums[y])

  else:

    pass


for z in range(len(sums)+1):

  incl = sumElemArr(sums,z,eraseZero(sums))
  arrSum.append(incl)
  #print(arrSum[z])


for w in range(len(arrSum)):

  if w < (len(arrSum)-1):

    l = arrSum[w][1] + arrSum[w+1][0]

    if l == 10:

      lni = w - 2
      resF[lni] += 1
      l = 0

    else:

      pass
      #print(l)

    resF.append(l)

  else:

    resF.append(arrSum[w][1])

#print(sums)

print("___________________________________ + \n")
printFormat(resF)

```
